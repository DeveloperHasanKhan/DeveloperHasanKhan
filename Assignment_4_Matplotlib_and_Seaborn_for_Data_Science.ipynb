```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the Gym data in google drive

from google.colab import drive
drive.mount('/content/drive')
import warnings
warnings.filterwarnings
gym = pd.read_csv('/content/drive/MyDrive/datasets1/gym_membership.csv')

```

```python
# Display last 5 rows

print("Display last 5 rows:\n", gym.tail())


# Check data types

print("Check data types: \n", gym.dtypes)
```

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

membership_types_count = gym['abonoment_type'].value_counts()
plt.bar(['Premium', 'Standar'],membership_types_count.values)
plt.title('Count membership types')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7,)
plt.savefig('/content/drive/MyDrive/datasets1/membership_types_count.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

plt.hist(gym['avg_time_in_gym'])
plt.title('time in gym')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7,)
plt.savefig('/content/drive/MyDrive/datasets1/avg_time_in_gym.png', dpi=300, bbox_inches='tight' )
plt.show()
```

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

from pickle import FALSE
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.grid(True)
plt.xlabel=('Age')
plt.ylabel=('visit per week')
plt.title=('Age vs Visit per week')
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig('/content/drive/MyDrive/datasets1/age_vs_visit_per_week.png', dpi=300, bbox_inches='tight' )
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')

# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

sns.boxplot(data=gym, x='gender', y='Age')
#plt.title('Age by Gender')
plt.savefig('/content/drive/MyDrive/datasets1/Gender_vs_Age.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
sample = gym.head(10)
sns.countplot(data=sample, x='days_per_week')
#plt.title("Days per week")
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.xticks(rotation=45)
plt.savefig("/content/drive/MyDrive/datasets1/count_days_per_week.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
#plt.title('Average visit_per_week Score by membership_type')
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/datasets1/visits.png",dpi=300, bbox_inches='tight')
plt.show()
```
